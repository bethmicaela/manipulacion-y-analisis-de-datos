**El Web Scraping ("raspado web" o "scrapear")** es un proceso dentro del Data Science que se utiliza para la extracción de datos de sitios web, normalmente en el lenguaje de programación Python y mediante machine learning, simulando cómo navegaría un ser humano en determinadas páginas web.

**El método de agrupación en conjunto k-means** es una técnica de aprendizaje automático no supervisado que se utiliza para identificar grupos de objetos de datos en un conjunto de datos

**Para este análisis de las letras en específico, es un discurso de un político de la ciudad de México**

```python
URL = 'http://lopezobrador.org.mx/2020/04/05/discurso-del-presidente-andres-manuel-lopez-obrador-en-su-informe-al-pueblo-de-mexico/'
website_url = rq.get(URL)
```

Con **website_url = rq.get(URL)** obtengo el html de la URL

```python
# Import BeautifulSoup from html structure information from requests (Importar BeautifulSoup desde la información de la estructura html de las solicitudes)
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url.text,'lxml')

data = [element.text for element in soup.find_all('p')] #Restricting text from 'p' founds
data = str(data) #Convert to String
data
```

**Beautiful Soup** es una biblioteca que facilita extraer información de páginas web. Se asienta sobre un analizador HTML o XML, proporcionando modismos Pythonic para iterar, buscar y modificar el árbol de análisis.

**Palabra vacía (Stop word)**

Palabras vacías es el nombre que reciben las palabras sin significado como artículos, pronombres, preposiciones, etc. que son filtradas antes o después del procesamiento de datos en lenguaje natural.

```python
from stop_words import get_stop_words # The word we want to ignore (La palabra que queremos ignorar)

stop_word_es = get_stop_words('es') # Spanish Stopwords (Palabras vacías en español)
```

```python
from wordcloud import WordCloud, STOPWORDS

print('wordcloud is installed and imported')
```

```python
stop_words = ['año','mil','millones','pesos','cientos','El','En'] + stop_word_es # Adding additional Stopwords

#Initiate world cloud  (Iniciar objeto de nube mundial)
amlo_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stop_words
)
# Generate de word cloud
amlo_wc.generate(data)
```

**amlo** es la página desde donde extraigo la información

```python
# Display the word cloud (mostrar la nube)
plt.imshow(amlo_wc, interpolation='bilinear')
plt.axis('off') # trazar imagen sin ejes
plt.show()
```

**La función imshow()**
en el módulo pyplot de la biblioteca matplotlib se usa para mostrar datos como una imagen; es decir, en un ráster regular 2D.

```python
# re-generate the word cloud
amlo_wc.generate(data)

# display the cloud
fig = plt.figure()
fig.set_figwidth(14) # set width (establecer ancho)
fig.set_figheight(18) # set height (establecer alto)

plt.imshow(amlo_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
