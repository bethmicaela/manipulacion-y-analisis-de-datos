## ¿que va a hacer nuestra funcion?

va a tomar una funcion lineal, las funciones lineales son muy faciles de identificar.

son de esta forma $$f(x)=mx + b$$

Tiene una pendiente, esto significa cuantos avanzamos en subida o ya sea en bajada a medida que damos otro paso.

Es como si pusieramos un carro vs una persona. Si esa persona puede recorrer 1 metro por segundo y si nuestro carro puede recorrer 10 metros por segundo.

Entonces cada paso es cada segundo que pasa.

La persona avanza 1 metro y el carro 10 metros, es decir que el carro se mueve más rapido.


$$m$$

significa cuanto avanzamos (es decir si subimos o bajamos) a medida que pasa el tiempo.

Entonces, si por cada segundo en el carro avanzamos 10 metros, entonces al pasar 10 segundos habrá avanzado 100 metros.


$$10*10$$ pero si lo resumimos nos daria como

$$10x$$ ->  donde x es la cantidad de segundos que han pasado.

Y esta tasa de crecimiento o de decrecimiento segun el caso.

y a esto se le conoce como la pendiente.

```
# Esta es nuestra funcion
def funcion_lineal(m, x, b=0):
  return m*x + b
```

```
# Creamos nuestros datos

# Creamos nuestros datos
N = 100

pendiente = 10

avance_carro = np.linspace(0, 20.0, num=N)

b = 0

y = funcion_lineal(pendiente, avance_carro, b)
```

```

fi, ax = plt.subplots()

# limitar la escala del eje x y el eje y.

# Esta escala es el rango o sea desde que numero hasta que numero, podemos representar
# lo mismo para la x, como el rango está en -10 a 10 para las x, solo vemos esos numeros, igual para las y pero es -100 a 100.

# si quisieramos cambiarlo, simplemente le asignamos numeros diferentes.
plt.xlim(0, 20)
plt.ylim(0, 200)

ax.plot(avance_carro, y)
ax.grid()

```

# Funciones seccionadas

son funciones que no pueden ser expresadas como poliniomios ejemplo:

### funciones trigonometricas

Dentro de las funciones trigonometricas tenemos a $sin(x)$,$cos(x)$ y a $tan(x)$

```
def funcion(x):
  return np.cos(x)

n = 1000

numeros = np.linspace(-2*np.pi, 2*np.pi, num=n)

y = funcion(numeros)


tablero = figure(title='Grafica del Coseno', x_axis_label='x', y_axis_label='y')

tablero.line(numeros, y)

show(tablero)
```

### Funciones exponenciales

$$f(x)=a^x$$
Es cuando elevamos un numero a otro numero, es decir multiplicar el numero a por si mismo por una cantidad x.

Ejemplo
$$f(5) = 2^5 = 2*2*2*2*2= 32 $$



```
def funcion_exponencial(x):
  return 2**x

n = 100
numeros = np.linspace(-10, 10 , num=n)

y = funcion_exponencial(numeros)
print(y)

lienzo = figure(title='Funcion Exponencial', x_axis_label='x', y_axis_label='y')

lienzo.line(numeros, y)

show(lienzo)
```

lo mismo para el numero de euler.


```
def funcion_e(x):
  return np.e **x

n = 100

x = np.linspace(0, 10, num=n)

y = funcion_e(x)

lienzo = figure(title='Funcion Exponencial', x_axis_label='x', y_axis_label='y')

lienzo.line(x,y, line_width=2, color='#CF3C83')

show(lienzo)
```

### Función logaritmo

El logaritmo está definido por la **realación**:

$$log_{b}(x) = n \Longleftrightarrow x=b^n$$

donde:



*   $b$ es la base.
*   $n$ es el exponente al que está elevado la base.
*   $x$ es el resultado de elevar la base $b$ al exponente $n$

**Ejemplo:**

Teniendo b=2 y n=8, entonces:

$$2^8=256$$

Por lo que $x=256$. Calculando el logaritmo base 2 de $x$ es:

$$log_{2}(256) = 8$$

Es decir le ponemos un numero al que queremos conocer el exponente al que queremos elevar, y este nos regresará el numero que es el exponente.




```
def funcion_logaritmica(x):
    return np.log2(x)

n = 100
x = np.linspace(0.01, 2000, num=n)

y = funcion_logaritmica(x)

figura = figure(title='Función Logaritmica', x_axis_label='X (resultados finales)', y_axis_label='Y (Exponentes)', x_range=[-500, 2000], y_range=[-10, 11])
figura.line(x,y, line_width=2, color='#513CCF')

show(figura)
```

### Funcion seccionada

Son funciones, que cambian de valor por intervalos, es como si dijeramos que llegamos a un punto y la maquina cambia su forma de procesar.

Ejemplo la funcion de Heaviside (Funcion por pasos)

$$H(x) =
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```
def heavy_side(x):
  a = np.zeros(len(x))

  for i, item in enumerate(x):
    if item > 0:
      a[i] = 1

  return a


n = 100

x = np.linspace(-10, 10, num=n)

y = heavy_side(x)

tablero = figure(title='Hevy Side', x_axis_label='X', y_axis_label='y', x_range=[-10, 10], y_range=[-1, 2])

tablero.line(x, y, line_width=2)

show(tablero)
```

# composicion de funciones


la composicion de funciones es que metemos lo que obtenemos de resultado en una funcion, lo pasamos como entrada a otra función.


```
def f(x):
  return 10*x

def g(x):
  return 2*x


numeros = np.linspace(-10.0, 10.0, num=1000)

f_o_g = g(g(numeros))
plt.plot(numeros, f_o_g)
```

# Funciones de activacion

Las funciones de activacion, son otro paso que se le hace a cada neurona, estas funciones nos dan la posibilidad de añadir mas flexibilidad a como se comportan los modelos matematicos dentro de cada neurona.

En principio, nos valdremos de la composicion de funciones, donde podemos imaginar algo asi:

* Una maquina que pela naranjas es una funcion, y una maquina que exprime naranjas es otra funcion distintas.

* El dato inicial es un bulto de naranjas

Entonces le pasamos la naranja a la primera maquina, que nos regresará una naranja pelada, despues esa naranja pelada se la pasamos a la otra maquina que nos regreserá un jugo de naranja.


# Funciones de activacion mas usadas

* Sigmoide
* Heaviside
* Tangente hiperbolica
* RELU

```
# Calculando relu

x = np.linspace(-10, 10, num=100)

def RELU(x):
  resultante = []

  for i in x:
    if i < 0:
      resultante.append(0)
    else:
      resultante.append(i)

  return np.array(resultante)


y = RELU(x)

plt.plot(x, y, color='G')
```
